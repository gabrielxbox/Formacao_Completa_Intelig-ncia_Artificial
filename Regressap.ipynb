```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import scipy.stats as stats
import seaborn as sns
```

```python
base = pd.read_csv('mt_cars.csv')
base.shape
```

```python
base.head()
```

```python
base = base.drop('Unnamed: 0', axis = 1)
```

```python
corr = base.corr()
sns.heatmap(corr, cmap='coolwarm', annot=False, fmt='.2f')
```

```python
column_parirs = [('mpg','cyl'),('mpg','disp'),('mpg','hp'),('mpg','wt'),('mpg','drat'),('mpg','vs')]
```

```python
n_plots = len(column_parirs)
fig, axes = plt.subplots(nrows=n_plots, ncols=1, figsize=(6,4 * n_plots))

for i, pair in enumerate(column_parirs):
    x_col, y_col = pair
    sns.scatterplot(x=x_col, y=y_col, data=base, ax=axes[i])
    axes[i].set_title(f'{x_col} vs  {y_col}')

plt.tight_layout()
plt.show()
```

```python
modelo = sm.ols(formula='mpg ~  wt + disp + hp ', data=base)
modelo = modelo.fit()
modelo.summary()
```

```python
residuos = modelo.resid
plt.hist(residuos, bins=20)
plt.xlabel('Residuos')
plt.ylabel('Frequncia')
plt.title('Histograma dos Residuos')
plt.show()
```

```python
stats.probplot(residuos, dist='norm', plot=plt)
plt.title('QQ Plot dos Residuos')
plt.show()
```

```python
start, pval = stats.shapiro(residuos)
print(f'Shapiro-wilk statstica: {start:.3f}, p-value: {pval:.3f}')
```
